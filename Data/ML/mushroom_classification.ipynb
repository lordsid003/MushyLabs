## Model Training (Mathematical modelling)

```python
def sigmoid(x):
    return (1 / (1 + np.exp(-x)))
```

```python
example_vector = np.arange(-10, 11)
example_vector = sigmoid(example_vector)
plt.plot(example_vector)
plt.title("Sigmoid Activation Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

```python
def compute_cost(x, y, w, b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    loss = 0.0
    for i in range(m):
        f_wb[i] = np.dot(w, x[i]) + b
        f_wb[i] = sigmoid(f_wb[i])
        loss += - (y[i] * np.log(f_wb[i])) - ((1 - y[i]) * np.log(1 - f_wb[i]))
    loss /= (2 * m)
    return loss
```

```python
def compute_gradient(x, y, w, b):
    m, n = x.shape
    f_wb = np.zeros(m)
    dj_dw = np.zeros(n)
    dj_db = 0.0
    error = 0.0
    for i in range(m):
        f_wb[i] = np.dot(w, x[i]) + b
        f_wb[i] = sigmoid(f_wb[i])
        error = (f_wb[i] - y[i])
        for j in range(n):
            dj_dw[j] += (error * x[i, j])
        dj_db += error
    dj_dw /= m
    dj_db /= m
    return dj_dw, dj_db
```

```python
import copy, math
```

```python
def gradient_descent(x, y, w_in, b_in, cost_func, gradient_func, alpha, num_iters):
    w = copy.deepcopy(w_in)
    b = b_in
    j_history = []
    w_history = []
    b_history = []
    for i in range(num_iters):
        dj_dw, dj_db = gradient_func(x, y, w, b)
        w -= alpha * dj_dw
        b -= alpha * dj_db
        if i < 100000:
            loss = cost_func(x, y, w, b)
            j_history.append(loss)
        if((i % math.ceil(num_iters / 10) == 0) or (i == num_iters - 1)):
            w_history.append(w)
            b_history.append(b)
            print(f"Iterations: {i} ===> Loss: {j_history[-1]}")
    return w, b, j_history, w_history, b_history
```

```python
w_initial = np.zeros((1, x_train.shape[1]))
b_initial = 0.01
alpha = 1e-2
iterations = 50000

w_final, b_final, loss, weights, bias = gradient_descent(x_train, y_train, w_initial, b_initial, compute_cost,
                                                         compute_gradient, alpha, iterations)
print(f"Trained Parameters:")
print(f"Weights: {w_final.tolist()}, Bias: {b_final}")
```

## Model Evaluation

```python
fig, ax = plt.subplots(1, 3, figsize=(20, 5))
plots = [loss, weights[0][0], bias]
labels = ["Loss", "Weights", "Bias"]
for i in range(3):
    ax[i].plot(plots[i])
    ax[i].set_title(f"{labels[i]} vs Epochs")
    ax[i].set_xlabel("Epochs")
    ax[i].set_ylabel(labels[i].upper())
plt.subplots_adjust(wspace=0.2)
plt.suptitle("Metrics Evaluation", fontsize=15)
plt.show()
```

## Accuracy on Training Data

```python
def prediction_matrix(x, w=w_final, b=b_final):
    m = x.shape[0]
    p = np.zeros(m)
    for i in range(m):
        p[i] = np.dot(w, x[i]) + b
        p[i] = sigmoid(p[i])
        if p[i] >= 0.5:
            p[i] = 1
        else:
            p[i] = 0
    return p
```

```python
def accuracy(x, y):
    m = x.shape[0]
    acc = 0.0
    for i in range(m):
        if x[i] == y[i]:
            acc += 1
    acc /= m
    return acc
```

```python
x_train_predictions = prediction_matrix(x_train)
x_train_accuracy = accuracy(x_train_predictions, y_train)
print(f"Training set accuracy: {x_train_accuracy * 100}%")
```

## Making predictions

```python
def predict(x, w=w_final, b=b_final):
    p = np.dot(w, x) + b
    p = sigmoid(p)
    if p >= 0.5:
        p = 1
    else:
        p = 0
    return p
```
